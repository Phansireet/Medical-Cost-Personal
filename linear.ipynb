```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import csv
```

```python
# Read data from CSV file
with open('insurance.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
data = np.array(data[1:])  # skip first row (header)
```

```python
# Extract bmi and charges as x and y values
x = data[:, 2].astype(np.float64)
y = data[:, 6].astype(np.float64)
```

```python
# Calculate linear regression
slope, intercept = np.polyfit(x, y, 1)
line = slope * x + intercept
```

```python
# Plot data points and line of best fit
plt.scatter(x, y)
plt.plot(x, line, color='red')

# Calculate integral and shade region under curve
def f(x):
    return slope * x + intercept
    
a = 20
b = 30
y_min = min(y)
integral, _ = quad(f, a, b)
x_integral = np.linspace(a, b, 100)
y_integral = f(x_integral)
plt.fill_between(x_integral, y_min, y_integral, where=(y_integral > y_min), color='gray')

# Show plot and integral value
plt.show()
print(f"The integral is {integral:.2f}")
print(f"y = {slope:.2f}x + {intercept:.2f}")
```
